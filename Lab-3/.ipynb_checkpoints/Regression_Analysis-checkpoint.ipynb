```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 

# Load data
data = pd.read_csv("house-prices.csv")
df = pd.DataFrame(data)

# Features and target variable
X = df[["SqFt"]]
y = df[["Price"]]

print(X, y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Plotting
plt.scatter(X, y, color="blue", label="Data Points")  
plt.plot(X, model.predict(X), color="red", label="Best Fit Line")  
plt.xlabel("SqFt")
plt.ylabel("Price")
plt.title("Linear Regression: Best Fit Line")
plt.legend()
plt.show()

# New prediction
new_data = pd.DataFrame([[60]], columns=["SqFt"])  # Create a DataFrame with the same column name
predicted_output = model.predict(new_data)

# Convert the output to a scalar before formatting
predicted_value = predicted_output[0].item()  # Extract scalar value from array
print(f"Predicted output for {new_data['SqFt'][0]}: {predicted_value:.2f}")
```
