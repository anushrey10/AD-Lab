```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

date_range = pd.date_range(start="2024-12-01", end="2024-12-31", freq='D')
np.random.seed(42)
temperatures = np.random.normal(loc=25, scale=5, size=len(date_range))  # Generate random temperatures

data = {
    "Date": date_range,
    "Temperature": temperatures
}
df = pd.DataFrame(data)
df.set_index("Date", inplace=True)

print("Time-Series Dataset:")
print(df.head())

rolling_window = 7  
df["7-Day Rolling Avg"] = df["Temperature"].rolling(window=rolling_window).mean()

plt.figure(figsize=(12, 6))

plt.plot(df.index, df["Temperature"], label="Daily Temperature", color='blue', linestyle='-', marker='o')

plt.plot(df.index, df["7-Day Rolling Avg"], label=f"{rolling_window}-Day Rolling Avg", color='orange', linestyle='-')

plt.title("Daily Temperatures and Rolling Average (December 2024)")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.xticks(rotation=45)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
