```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
marks1 = np.random.randint(0,101,size=20)
marks2 = marks1+np.random.randint(-10,10,size=marks1.size)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(marks1, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Marks (Subject 1)")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.scatter(marks1, marks2, color='green', marker='o')
plt.title("Scatter Plot: Subject 1 vs Subject 2")
plt.xlabel("Subject 1 Marks")
plt.ylabel("Subject 2 Marks")
plt.axline((0, 0), slope=1, color='red', linestyle='--', label="y=x (Perfect Correlation)")
plt.legend()

plt.tight_layout()
plt.show()
```
