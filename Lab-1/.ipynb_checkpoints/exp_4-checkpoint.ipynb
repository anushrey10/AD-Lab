```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    "Month": [
        "January", "February", "March", "April", "May", "June",
        "July", "August", "September", "October", "November", "December"
    ],
    "Sales": [50, 60, 65, 70, 80, 85, 90, 95, 110, 120, 130, 140]
}

df = pd.DataFrame(data)
print("Sales Data:")
print(df)

sales = np.array(df["Sales"])
growth_rates = np.zeros(len(sales))
growth_rates[1:] = (sales[1:] - sales[:-1]) / sales[:-1] * 100  # Skip the first month
df["Growth Rate (%)"] = growth_rates

print("\nSales Data with Growth Rates:")
print(df)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(df["Month"], df["Sales"], marker='o', color='blue', label="Sales")
plt.title("Monthly Sales Data")
plt.xlabel("Month")
plt.ylabel("Sales (in $1000)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.bar(df["Month"], df["Growth Rate (%)"], color='orange', label="Growth Rate")
plt.title("Monthly Growth Rates")
plt.xlabel("Month")
plt.ylabel("Growth Rate (%)")
plt.xticks(rotation=45)
plt.axhline(0, color='red', linestyle='--')  # Line for zero growth
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
